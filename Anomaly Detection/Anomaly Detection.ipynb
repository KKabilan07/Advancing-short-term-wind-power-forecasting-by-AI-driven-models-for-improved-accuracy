Anomaly Detection

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from tensorflow.keras.layers import LSTM, Dense, Input, Lambda, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
from sklearn.model_selection import train_test_split, KFold
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import STL
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_recall_fscore_support
import pywt

# Load and preprocess data
df = pd.read_csv("/kaggle/input/wind-turbine-scada-dataset/T1.csv", parse_dates=['Date/Time'], dayfirst=True)
df = df.sort_values('Date/Time')

# Step 1: Updated Physics-Informed Preprocessing
def calculate_power_curve(wind_speed, theoretical_power, active_power):
    """Updated physical constraint: Check for extreme wind speeds or power exceeding theoretical max."""
    violation_mask = ((wind_speed > 25) | (active_power > 1.2 * theoretical_power)) & (wind_speed > 3)
    return violation_mask.astype(int)

df['power_violation'] = calculate_power_curve(df['Wind Speed (m/s)'], df['Theoretical_Power_Curve (KWh)'], df['LV ActivePower (kW)'])

# Step 2: Anomaly Detection Model (LSTM Autoencoder) with Updated detect_anomalies
class WindPowerAnomalyDetector:
    def __init__(self, time_steps=6, features=4):
        self.time_steps = time_steps
        self.features = features
        self.scaler = RobustScaler()

        # Autoencoder architecture
        inputs = Input(shape=(time_steps, features))
        encoded = LSTM(32, activation='tanh', return_sequences=True)(inputs)
        encoded = LSTM(16, activation='tanh')(encoded)

        decoded = Lambda(lambda x: K.repeat_elements(K.expand_dims(x, 1), time_steps, 1))(encoded)
        decoded = LSTM(16, activation='tanh', return_sequences=True)(decoded)
        decoded = LSTM(32, activation='tanh', return_sequences=True)(decoded)
        decoded = Dense(features)(decoded)

        self.autoencoder = Model(inputs, decoded)
        self.autoencoder.compile(optimizer='adam', loss='mse')

    def create_sequences(self, data):
        X = []
        for i in range(len(data) - self.time_steps):
            X.append(data[i:i + self.time_steps])
        return np.array(X)

    def fit(self, data, epochs=20, batch_size=32, validation_split=0.1):
        scaled_data = self.scaler.fit_transform(data)
        sequences = self.create_sequences(scaled_data)
        history = self.autoencoder.fit(sequences, sequences,
                                      epochs=epochs,
                                      batch_size=batch_size,
                                      shuffle=False,
                                      validation_split=validation_split,
                                      verbose=1)
        return history

    def detect_anomalies(self, data, threshold_percentile=95):
        scaled_data = self.scaler.transform(data)
        sequences = self.create_sequences(scaled_data)
        reconstructions = self.autoencoder.predict(sequences)
        mse = np.mean(np.square(sequences - reconstructions), axis=(1, 2))
        threshold = np.percentile(mse, threshold_percentile)
        anomalies = mse > threshold
        return anomalies, mse, threshold

# Step 3: Physics-Informed Neural Network
# Step 3: Physics-Informed Neural Network
class PhysicsInformedWindModel(tf.keras.Model):
    def __init__(self, hidden_layers=[64, 32], dropout_rate=0.2):
        super().__init__()
        self.dense_layers = [
            Dense(units, activation='tanh', kernel_regularizer=tf.keras.regularizers.l2(0.01))
            for units in hidden_layers
        ]
        self.dropout_layers = [Dropout(dropout_rate) for _ in hidden_layers]
        self.final_layer = Dense(1)

        # Physical parameters (wind turbine characteristics)
        self.rho = 1.225  # Air density (kg/m³)
        self.A = 7854.0   # Rotor area (m²) for 100m diameter
        self.Cp = 0.4     # Power coefficient

    def call(self, inputs, training=False):
        x = inputs
        for i, layer in enumerate(self.dense_layers):
            x = layer(x)
            if training:
                x = self.dropout_layers[i](x)
        return self.final_layer(x)

    def physics_loss(self, wind_speed, predictions):
        # Betz' Law power calculation (physical constraint)
        theoretical_power = 0.5 * self.rho * self.A * self.Cp * (wind_speed**3)
        return tf.reduce_mean(tf.square(predictions - theoretical_power))

    def train_step(self, data):
        inputs, targets = data
        wind_speed = inputs[:, 0]  # Assuming wind speed is the first feature

        with tf.GradientTape() as tape:
            predictions = self(inputs, training=True)
            data_loss = self.compiled_loss(targets, predictions)
            physics_loss = self.physics_loss(wind_speed, predictions)
            total_loss = data_loss + 0.1 * physics_loss  # Weighted sum

        self.compiled_metrics.update_state(targets, predictions)
        gradients = tape.gradient(total_loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))
        return {m.name: m.result() for m in self.metrics}

# Step 4: Integrated Implementation with K-Fold Cross-Validation
features = ['Wind Speed (m/s)', 'Wind Direction (°)', 'Theoretical_Power_Curve (KWh)', 'LV ActivePower (kW)']
data = df[features].values
# K-Fold Cross-Validation for Anomaly Detection
kf = KFold(n_splits=5, shuffle=True, random_state=42)
fold_metrics = []

for fold, (train_idx, test_idx) in enumerate(kf.split(data)):
    print(f"Processing Fold {fold + 1}...")
    train_data = data[train_idx]
    test_data = data[test_idx]

    # Initialize and train the anomaly detector
    anomaly_detector = WindPowerAnomalyDetector(time_steps=6, features=4)
    history = anomaly_detector.fit(train_data, epochs=20, batch_size=32, validation_split=0.1)

    # Detect anomalies on the test set
    anomalies_test, mse_test, threshold = anomaly_detector.detect_anomalies(test_data, threshold_percentile=95)

    # Adjust anomalies array for test set
    anomalies_test_full = np.concatenate([anomalies_test, np.full(anomaly_detector.time_steps, False)])

    # Ground truth for test set
    ground_truth_test = df['power_violation'].iloc[test_idx].values
    ground_truth_test = ground_truth_test[:-anomaly_detector.time_steps]  # Align with anomalies_test

    # Confusion Matrix
    cm = confusion_matrix(ground_truth_test, anomalies_test)
    precision, recall, f1, _ = precision_recall_fscore_support(ground_truth_test, anomalies_test, average='binary', zero_division=0)
    fold_metrics.append((precision, recall, f1))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal', 'Anomaly'])
    disp.plot()
    plt.title(f'Confusion Matrix for Fold {fold + 1}')
    plt.savefig(f'/kaggle/working/confusion_matrix_fold_{fold + 1}.png')
    plt.show()

    print(f"Fold {fold + 1} - Precision: {precision:.2f}, Recall: {recall:.2f}, F1-Score: {f1:.2f}")

# Average metrics across folds
avg_precision, avg_recall, avg_f1 = np.mean(fold_metrics, axis=0)
print(f"Average Precision: {avg_precision:.2f}, Average Recall: {avg_recall:.2f}, Average F1-Score: {avg_f1:.2f}")

# Proceed with PINN training on clean data (excluding anomalies from the entire dataset)
anomalies, mse_values, threshold = anomaly_detector.detect_anomalies(data, threshold_percentile=95)
anomalies = np.concatenate([anomalies, np.full(anomaly_detector.time_steps, False)])

X = data[:-anomaly_detector.time_steps, :-1]
y = data[:-anomaly_detector.time_steps, -1]
anomalies = anomalies[:-anomaly_detector.time_steps]

clean_X = X[~anomalies]
clean_y = y[~anomalies]

# Save clean dataset
clean_df = df.iloc[:-anomaly_detector.time_steps].copy()
clean_df = clean_df[~anomalies]
clean_df.to_csv('/kaggle/working/clean_data.csv', index=False)
print("Clean dataset saved to '/kaggle/working/clean_data.csv'")

# Split the clean data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(clean_X, clean_y, test_size=0.2, shuffle=False)

# Train PINN model
pinn_model = PhysicsInformedWindModel(dropout_rate=0.3)
pinn_model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss='mse', metrics=['mae'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = pinn_model.fit(X_train, y_train, epochs=100, batch_size=128,
                        validation_data=(X_val, y_val),
                        callbacks=[early_stopping],
                        verbose=1)

# Step 5: Physics-Constrained Prediction
def predict_with_constraints(model, input_data):
    wind_speed = input_data[:, 0]
    raw_pred = model.predict(input_data)
    theoretical_max = 0.5 * 1.225 * 7854 * 0.4 * (wind_speed**3)
    return np.minimum(raw_pred, theoretical_max)

new_data = np.array([[7.5, 245, 1200], [2.0, 180, 50]])
predictions = predict_with_constraints(pinn_model, new_data)
print("Predictions with Physics Constraints:")
print(predictions)

# Step 6: Visualization and Analysis
def visualize_results(df, anomalies, history):
    plt.figure(figsize=(12, 6))
    plt.plot(df['Date/Time'], df['LV ActivePower (kW)'], label='Power Output')
    anomalies_plot = np.concatenate([anomalies, np.full(len(df) - len(anomalies), False)])
    plt.scatter(df['Date/Time'][anomalies_plot], df['LV ActivePower (kW)'][anomalies_plot],
                c='red', label='Anomalies')
    plt.xlabel('Time', fontsize="20")
    plt.ylabel('Power (kW)', fontsize="20")
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(fontsize=12, loc="upper right")
    plt.savefig('/kaggle/working/ad1.png')
    plt.show()

visualize_results(df, anomalies, history)

def interactive_time_series(df):
    fig = go.Figure(data=[go.Scatter(x=df['Date/Time'], y=df['LV ActivePower (kW)'])])
    fig.update_layout(title='Interactive Wind Power Output', xaxis_title='Time', yaxis_title='Power (kW)')
    fig.show()

interactive_time_series(df)

def heatmap_correlation(df):
    corr_matrix = df.drop(columns=['Date/Time']).corr()
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
    plt.savefig('/kaggle/working/hm.png')
    plt.show()

heatmap_correlation(df)

def scatter_plot_anomalies(df, anomalies):
    plt.figure(figsize=(10, 6))
    num_data_points = len(df)
    anomalies_plot = anomalies[:num_data_points]
    if len(anomalies_plot) < num_data_points:
        anomalies_plot = np.concatenate([anomalies_plot, np.full(num_data_points - len(anomalies_plot), False)])
    plt.scatter(df['Wind Speed (m/s)'], df['LV ActivePower (kW)'], c=['blue' if not a else 'red' for a in anomalies_plot])
    plt.xlabel('Wind Speed (m/s)', fontsize="20")
    plt.ylabel('Power (kW)', fontsize="20")
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.savefig('/kaggle/working/ad2.png')
    plt.show()

scatter_plot_anomalies(df, anomalies)

def apply_wavelet_decomposition(data, wavelet='db4', level=3):
    coeffs = pywt.wavedec(data, wavelet, level=level)
    return coeffs

def visualize_wavelet_decomposition(df):
    plt.figure(figsize=(12, 8))
    signal = df['LV ActivePower (kW)'].values
    coeffs = apply_wavelet_decomposition(signal)
    
    plt.subplot(len(coeffs) + 1, 1, 1)
    plt.plot(signal, label='Original Signal')
    plt.title('Original Signal')
    
    for i, coeff in enumerate(coeffs):
        plt.subplot(len(coeffs) + 1, 1, i + 2)
        plt.plot(coeff, label=f'Level {i} Coefficient')
        plt.title(f'Wavelet Coefficient Level {i}')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/wavelet_decomposition.png')
    plt.show()

visualize_wavelet_decomposition(df)

def plot_stl_decomposition(df):
    stl = STL(df['LV ActivePower (kW)'], period=144)
    res = stl.fit()
    fig, axes = plt.subplots(ncols=1, nrows=4, sharex=True, figsize=(12, 8))
    
    axes[0].plot(res.observed)
    axes[0].set_ylabel('Power (kW)')
    
    axes[1].plot(res.trend)
    axes[1].set_ylabel('Trend Component')
    
    axes[2].plot(res.seasonal)
    axes[2].set_ylabel('Seasonal Component')
    
    axes[3].plot(res.resid)
    axes[3].set_ylabel('Residual Component')
    
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('/kaggle/working/stl.png')
    plt.show()

plot_stl_decomposition(df)

print("Betz' Law is incorporated into the PINN loss function as follows:")
print("Theoretical power = 0.5 * rho * A * Cp * (wind_speed^3)")
print("Physics Loss = Mean Squared Error between predictions and theoretical power")
print("Total Loss = Data Loss + 0.1 * Physics Loss")
print("The weight (0.1) of the physics loss is tunable and can be adjusted to balance the influence of physical constraints.")
```
