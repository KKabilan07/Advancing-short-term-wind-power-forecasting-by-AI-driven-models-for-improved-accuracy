Temporal Fusion Transformer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Preprocessing and evaluation tools
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Deep learning imports
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

#############################################
# 1. Data Loading and Feature Engineering
#############################################
file_path = "/kaggle/input/eeefinal/T2.csv"  # Update this path as needed
df = pd.read_csv(file_path)

# Convert the 'Date/Time' column to datetime and set as index
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%d %m %Y %H:%M', dayfirst=True)
df.set_index('Date/Time', inplace=True)

# Use the entire dataset (e.g. ~50,531 instances)
print("Total instances available:", len(df))

# Before any transformations, save original target values for later inversion
target_col = df.columns[-1]
original_target_values = df[target_col].values
target_min = original_target_values.min()
target_max = original_target_values.max()
print(f"Target range: [{target_min}, {target_max}]")

# Feature enrichment:
# Add hour and day-of-week and transform them to cyclic features:
df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['dow_sin']  = np.sin(2 * np.pi * df['dayofweek'] / 7)
df['dow_cos']  = np.cos(2 * np.pi * df['dayofweek'] / 7)

# Create a lag feature for the target
df['lag1'] = df[target_col].shift(1)

# Drop rows with missing values (from lag or otherwise)
df = df.dropna()
print("Data shape after enrichment:", df.shape)

#############################################
# 2. Improved Normalization and PCA
#############################################
# Separate features and target
X_df = df.drop(columns=[target_col])
y_df = df[target_col]

# Store target values for reference
target_values = y_df.values

# Scale features and target separately
X_scaler = StandardScaler()
y_scaler = StandardScaler()

X_scaled = X_scaler.fit_transform(X_df)
y_scaled = y_scaler.fit_transform(y_df.values.reshape(-1, 1)).flatten()

# Store these for inverse transformation later
target_mean = y_scaler.mean_
target_std = y_scaler.scale_
print(f"Target mean: {target_mean}, Target std: {target_std}")

# Apply PCA only to the features, not the target
pca = PCA(n_components=0.90, svd_solver='full')
X_pca = pca.fit_transform(X_scaled)
print("Shape after PCA (features only):", X_pca.shape)

# Combine PCA-transformed features with scaled target for sequence creation
pca_data = np.column_stack((y_scaled, X_pca))
print("Combined data shape:", pca_data.shape)

#############################################
# 3. Create Sliding Window Sequences with Index Tracking
#############################################
window_size = 50  # number of timesteps per sample

def create_sequences(data, window_size):
    X, y = [], []
    original_indices = []  # Track original indices for mapping back
    
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size, :])
        # Target is the first column (scaled target)
        y.append(data[i+window_size, 0])
        # Store original index
        original_indices.append(i+window_size)
        
    return np.array(X), np.array(y), np.array(original_indices)

X_seq, y_seq, seq_indices = create_sequences(pca_data, window_size)
print("Sequence shape:", X_seq.shape, "Target shape:", y_seq.shape)

#############################################
# 4. Train-Test Split with Index Tracking and Noise Augmentation
#############################################
X_train, X_test, y_train, y_test, train_indices, test_indices = train_test_split(
    X_seq, y_seq, seq_indices, test_size=0.2, random_state=42
)

# Store original target values corresponding to test indices for later comparison
original_y_test = target_values[test_indices]

# Add Gaussian noise to training data for augmentation
def add_gaussian_noise(data, noise_level=0.01):
    noise = noise_level * np.random.randn(*data.shape)
    return data + noise

X_train_noisy = add_gaussian_noise(X_train, noise_level=0.01)
print("Training shape after noise augmentation:", X_train_noisy.shape)

#############################################
# 5. Build an Improved Temporal Fusion Transformer (TFT) Model
#############################################
def build_improved_tft_model(input_shape, lstm_units=128, attn_heads=4, attn_dim=64, dropout=0.2):
    inputs = layers.Input(shape=input_shape)  # (window_size, features)
    
    # Extract the target feature from input sequences for skip connection
    target_feature = layers.Lambda(lambda x: x[:, :, 0])(inputs)
    target_feature = layers.Reshape((-1, 1))(target_feature)
    
    # Variable selection / feature transformation block
    x = layers.TimeDistributed(layers.Dense(128, activation='relu', 
                                           kernel_regularizer=regularizers.l2(1e-4)))(inputs)
    
    # LSTM encoder with increased units to capture temporal dependencies
    lstm_out = layers.LSTM(lstm_units, return_sequences=True,
                          kernel_regularizer=regularizers.l2(1e-4))(x)
    
    # Multi-head attention block over LSTM outputs with increased heads and dimensions
    attn_out = layers.MultiHeadAttention(num_heads=attn_heads, key_dim=attn_dim, dropout=dropout)(lstm_out, lstm_out)
    attn_out = layers.Dropout(dropout)(attn_out)
    
    # Residual connection
    x = layers.Add()([lstm_out, attn_out])
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    
    # Attention-based pooling
    attention = layers.Dense(1, activation='tanh')(x)
    attention = layers.Reshape((-1,))(attention)
    attention_weights = layers.Softmax()(attention)
    context = layers.Dot(axes=(1, 1))([x, layers.Reshape((-1, 1))(attention_weights)])
    context = layers.Flatten()(context)
    
    # Skip connection from input target features (averaged)
    avg_target = layers.GlobalAveragePooling1D()(target_feature)
    
    # Combine context with skip connection
    x = layers.Concatenate()([context, avg_target])
    
    # Final dense layers with stronger regularization
    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(dropout)(x)
    outputs = layers.Dense(1)(x)
    
    model = models.Model(inputs, outputs)
    
    # Use optimizer with learning rate schedule
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss="mse", metrics=["mae"])
    return model

input_shape = X_train_noisy.shape[1:]
model_tft_noise = build_improved_tft_model(input_shape)
model_tft_noise.summary()

#############################################
# 6. Training with Early Stopping and Learning Rate Reduction
#############################################
early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
lr_reduce = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, verbose=1)

history_tft_noise = model_tft_noise.fit(X_train_noisy, y_train,
                                        epochs=50,
                                        batch_size=32,
                                        validation_split=0.2,
                                        callbacks=[early_stop, lr_reduce],
                                        verbose=1)

#############################################
# 7. Evaluation with Proper Inverse Transformation
#############################################
y_pred_tft_noise = model_tft_noise.predict(X_test).flatten()

# Calculate metrics on the scaled data
mse_tft_noise = mean_squared_error(y_test, y_pred_tft_noise)
rmse_tft_noise = np.sqrt(mse_tft_noise)
mae_tft_noise = mean_absolute_error(y_test, y_pred_tft_noise)
r2_tft_noise = r2_score(y_test, y_pred_tft_noise)

print("\n[Improved TFT with Preprocessing] Performance Metrics on Scaled Data:")
print(f"MSE: {mse_tft_noise:.4f}")
print(f"RMSE: {rmse_tft_noise:.4f}")
print(f"MAE: {mae_tft_noise:.4f}")
print(f"R²: {r2_tft_noise:.4f}")

# ---------------------------
# 8. Inverse Transform Data for Plotting (ORIGINAL kW SCALE)
# ---------------------------
# Simple inverse transform: apply the inverse of the StandardScaler to get original kW values
def inverse_transform_target(scaled_y):
    # Convert to 2D array for inverse_transform
    scaled_y_2d = scaled_y.reshape(-1, 1)
    # Apply inverse transform to get original kW scale
    original_y = y_scaler.inverse_transform(scaled_y_2d).flatten()
    return original_y

# Inverse transform predictions and test values to original kW scale
original_scale_y_pred = inverse_transform_target(y_pred_tft_noise)
# original_y_test was already saved during train_test_split and is in original kW scale

print(f"TFT Predictions range (kW): [{original_scale_y_pred.min():.2f}, {original_scale_y_pred.max():.2f}]")
print(f"Test values range (kW): [{original_y_test.min():.2f}, {original_y_test.max():.2f}]")

# ---------------------------
# 9. Save Learning Curves Data to CSV
# ---------------------------
# Create DataFrame for learning curves
learning_curves_df = pd.DataFrame({
    'epoch': range(1, len(history_tft_noise.history["loss"]) + 1),
    'training_loss': history_tft_noise.history["loss"],
    'validation_loss': history_tft_noise.history["val_loss"]
})

# Save learning curves data to CSV
learning_curves_df.to_csv('learning_curves_TFT.csv', index=False)
print("Learning curves data saved to 'learning_curves_TFT.csv'")

# ---------------------------
# 10. Save Predictions vs True Values to CSV (Original kW Scale)
# ---------------------------
# Create DataFrame for predictions vs true values in original kW scale
predictions_df = pd.DataFrame({
    'sample_index': range(len(original_y_test)),
    'true_values_kW': original_y_test,
    'predictions_kW': original_scale_y_pred
})

# Save predictions data to CSV
predictions_df.to_csv('predictions_vs_true_TFT.csv', index=False)
print("Predictions vs true values data saved to 'predictions_vs_true_TFT.csv'")

# Plot learning curves with improved formatting
plt.figure(figsize=(10, 5))
plt.plot(history_tft_noise.history["loss"], label="Training Loss")
plt.plot(history_tft_noise.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs", fontsize="16")
plt.ylabel("MSE Loss", fontsize="20")
plt.legend()
plt.xticks(fontsize=14)
plt.yticks(fontsize=16)
plt.title("TFT Learning Curves", fontsize="16")
plt.savefig('/kaggle/working/learning_curvesTFT.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot predictions vs true values using original kW scale
plt.figure(figsize=(10, 5))
plt.plot(original_y_test[:1000], label="True Values", marker="o", markersize=2)
plt.plot(original_scale_y_pred[:1000], label="Predictions", marker="x", markersize=2)
plt.xlabel("Sample Index", fontsize="15.5")
plt.ylabel("Wind Power (kW)", fontsize="20")
plt.legend(loc='center left')
plt.xticks(fontsize=13)
plt.yticks(fontsize=16)
plt.title("TFT Predictions (Original kW Scale)", fontsize="16")
plt.savefig('/kaggle/working/predictions_vs_trueTFT.png', dpi=800, bbox_inches='tight')
plt.show()

# Calculate metrics on the original kW scale data
mse_orig = mean_squared_error(original_y_test, original_scale_y_pred)
rmse_orig = np.sqrt(mse_orig)
mae_orig = mean_absolute_error(original_y_test, original_scale_y_pred)
r2_orig = r2_score(original_y_test, original_scale_y_pred)

print("\n[TFT - Original kW Scale] Performance Metrics:")
print(f"MSE: {mse_orig:.4f} kW²")
print(f"RMSE: {rmse_orig:.4f} kW")
print(f"MAE: {mae_orig:.4f} kW")
print(f"R²: {r2_orig:.4f}")

# Additional TFT-specific analysis - Error distribution
error = original_y_test - original_scale_y_pred
print(f"\nTFT Error Analysis:")
print(f"Mean Error: {np.mean(error):.4f} kW")
print(f"Error Std: {np.std(error):.4f} kW")
print(f"Max Absolute Error: {np.max(np.abs(error)):.4f} kW")

# Error distribution plot (optional)
plt.figure(figsize=(8, 5))
plt.hist(error, bins=50, alpha=0.7, edgecolor='black')
plt.xlabel("Prediction Error (kW)", fontsize="14")
plt.ylabel("Frequency", fontsize="14")
plt.title("TFT Prediction Error Distribution", fontsize="16")
plt.axvline(x=0, color='red', linestyle='--', label='Perfect Prediction')
plt.legend()
plt.savefig('/kaggle/working/error_distribution_TFT.png', dpi=300, bbox_inches='tight')
plt.show()
```
