```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ---------------------------
# 1. Data Loading and Enrichment
# ---------------------------
file_path = '/kaggle/input/eeewindpower/T2.csv'
df = pd.read_csv(file_path)

# Convert and set datetime index
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%d %m %Y %H:%M', dayfirst=True)
df.set_index('Date/Time', inplace=True)

print("Total instances available:", len(df))

# Save original target values
target_col = df.columns[-1]
original_target_values = df[target_col].values
target_min = original_target_values.min()
target_max = original_target_values.max()
print(f"Target range: [{target_min}, {target_max}]")

# Feature engineering
df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['dow_sin'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
df['dow_cos'] = np.cos(2 * np.pi * df['dayofweek'] / 7)
df['lag1'] = df[target_col].shift(1)
df = df.dropna()
print("Data shape after enrichment:", df.shape)

# ---------------------------
# 2. Normalization and PCA
# ---------------------------
X_df = df.drop(columns=[target_col])
y_df = df[target_col]
target_values = y_df.values

X_scaler = StandardScaler()
y_scaler = StandardScaler()
X_scaled = X_scaler.fit_transform(X_df)
y_scaled = y_scaler.fit_transform(y_df.values.reshape(-1, 1)).flatten()

target_mean = y_scaler.mean_
target_std = y_scaler.scale_
print(f"Target mean: {target_mean}, Target std: {target_std}")

# PCA with variance analysis
pca = PCA(n_components=0.90, svd_solver='full')
X_pca = pca.fit_transform(X_scaled)
print("Shape after PCA:", X_pca.shape)

# Generate PCA scree plot
plt.figure(figsize=(10,6))
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, 'bo-', label='Cumulative Explained Variance')
plt.bar(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, alpha=0.6, color='g', label='Individual Variance')
plt.axhline(y=0.90, color='r', linestyle='--', label='90% Threshold')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title(f'PCA Scree Plot ({pca.n_components_} Components Explain 90% Variance)')
plt.legend()
plt.grid(True)
plt.savefig('pca_scree_plot.png', dpi=300)
plt.show()

# Save PCA variance data
pca_df = pd.DataFrame({
    'Component': range(1, len(pca.explained_variance_ratio_)+1),
    'Individual_Variance': pca.explained_variance_ratio_,
    'Cumulative_Variance': cumulative_variance
})
pca_df.to_csv('pca_variance_analysis.csv', index=False)

# Combine data
pca_data = np.column_stack((y_scaled, X_pca))
print("Combined data shape:", pca_data.shape)

# ---------------------------
# 3. Create Sequences
# ---------------------------
window_size = 50

def create_sequences(data, window_size):
    X, y, indices = [], [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size, :])
        y.append(data[i+window_size, 0])
        indices.append(i+window_size)
    return np.array(X), np.array(y), np.array(indices)

X_seq, y_seq, seq_indices = create_sequences(pca_data, window_size)
print("Sequence shape:", X_seq.shape, "Target shape:", y_seq.shape)

# ---------------------------
# 4. Train-Test Split
# ---------------------------
X_train, X_test, y_train, y_test, train_indices, test_indices = train_test_split(
    X_seq, y_seq, seq_indices, test_size=0.2, random_state=42
)
original_y_test = target_values[test_indices]

# Add Gaussian noise only to training data
def add_gaussian_noise(data, noise_level=0.01):
    noise = noise_level * np.random.randn(*data.shape)
    return data + noise

X_train_noisy = add_gaussian_noise(X_train, noise_level=0.01)
print("Training set shape after noise:", X_train_noisy.shape)

# ---------------------------
# 5. Transformer Model
# ---------------------------
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0.1):
    x = layers.LayerNormalization(epsilon=1e-6)(inputs)
    x = layers.MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(x, x)
    x = layers.Dropout(dropout)(x)
    res = layers.Add()([inputs, x])
    
    x = layers.LayerNormalization(epsilon=1e-6)(res)
    x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu",
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1,
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    return layers.Add()([res, x])

def build_transformer_model(input_shape, head_size=64, num_heads=4, ff_dim=256, num_layers=3, dropout=0.2):
    inputs = layers.Input(shape=input_shape)
    target_feature = layers.Lambda(lambda x: x[:, :, 0])(inputs)
    target_feature = layers.Reshape((-1, 1))(target_feature)
    
    x = inputs
    for _ in range(num_layers):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)
    
    attention = layers.Dense(1, activation='tanh')(x)
    attention = layers.Reshape((-1,))(attention)
    attention_weights = layers.Softmax()(attention)
    context = layers.Dot(axes=(1, 1))([x, layers.Reshape((-1, 1))(attention_weights)])
    context = layers.Flatten()(context)
    
    avg_target = layers.GlobalAveragePooling1D()(target_feature)
    x = layers.Concatenate()([context, avg_target])
    
    x = layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(64, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(dropout)(x)
    outputs = layers.Dense(1)(x)
    
    model = models.Model(inputs, outputs)
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss="mse", metrics=["mae"])
    return model

input_shape = X_train_noisy.shape[1:]
model_noise = build_transformer_model(input_shape)
model_noise.summary()

# ---------------------------
# 6. Training
# ---------------------------
early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
lr_reduce = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, verbose=1)

history_noise = model_noise.fit(
    X_train_noisy, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, lr_reduce],
    verbose=1
)

# ---------------------------
# 7. Evaluation
# ---------------------------
y_pred_noise = model_noise.predict(X_test).flatten()

# Inverse transform to original kW scale
def inverse_transform_target(scaled_y):
    return y_scaler.inverse_transform(scaled_y.reshape(-1, 1)).flatten()

original_scale_y_pred = inverse_transform_target(y_pred_noise)

# Calculate metrics
mse_orig = mean_squared_error(original_y_test, original_scale_y_pred)
rmse_orig = np.sqrt(mse_orig)
mae_orig = mean_absolute_error(original_y_test, original_scale_y_pred)
r2_orig = r2_score(original_y_test, original_scale_y_pred)

# MAPE with zero-handling
def calculate_mape(y_true, y_pred):
    non_zero_mask = y_true != 0
    if np.sum(non_zero_mask) == 0:
        return float('nan')
    absolute_percentage = np.abs((y_true[non_zero_mask] - y_pred[non_zero_mask]) / y_true[non_zero_mask])
    return np.mean(absolute_percentage) * 100

mape_val = calculate_mape(original_y_test, original_scale_y_pred)

print("\n[ORIGINAL SCALE - kW] Performance Metrics:")
print(f"MSE: {mse_orig:.4f}")
print(f"RMSE: {rmse_orig:.4f}")
print(f"MAE: {mae_orig:.4f}")
print(f"MAPE: {mape_val:.2f}%")
print(f"R: {r2_orig:.4f}")

# Save metrics
metrics_df = pd.DataFrame({
    'metric': ['MSE', 'RMSE', 'MAE', 'MAPE', 'R2'],
    'value': [mse_orig, rmse_orig, mae_orig, mape_val, r2_orig]
})
metrics_df.to_csv('performance_metrics_VT.csv', index=False)

# ---------------------------
# 8. Visualization
# ---------------------------
# Learning curves
plt.figure(figsize=(10, 5))
plt.plot(history_noise.history["loss"], label="Training Loss")
plt.plot(history_noise.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("MSE Loss", fontsize=14)
plt.title("Learning Curves", fontsize=16)
plt.legend()
plt.grid(True)
plt.savefig('learning_curves_VT.png', dpi=300)
plt.show()

# Save learning curve data
lc_df = pd.DataFrame({
    'epoch': range(1, len(history_noise.history["loss"]) + 1),
    'train_loss': history_noise.history["loss"],
    'val_loss': history_noise.history["val_loss"]
})
lc_df.to_csv('learning_curves_data_VT.csv', index=False)

# Prediction plot (DENORMALIZED - kW)
plt.figure(figsize=(12, 6))
plt.plot(original_y_test[:1000], 'b-', label="True Values", linewidth=1.5, alpha=0.8)
plt.plot(original_scale_y_pred[:1000], 'r--', label="Predictions", linewidth=1.5)
plt.xlabel("Sample Index", fontsize=14)
plt.ylabel("Wind Power (kW)", fontsize=14)
plt.title("Vanilla Transformer Predictions (Denormalized kW Scale)", fontsize=16)
plt.legend()
plt.grid(True)
plt.savefig('predictions_denormalized_kW_VT.png', dpi=300)
plt.show()

# Save prediction data
preds_df = pd.DataFrame({
    'true_kW': original_y_test,
    'pred_kW': original_scale_y_pred
})
preds_df.to_csv('predictions_denormalized_VT.csv', index=False)

print("All outputs saved successfully")
```
